```python
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data = load_digits()
```

```python
# 피처
digits_data = data.data

# 레이블, 타겟
digits_label = data.target
digits_label_names = data.target_names

print(digits_data)
print('=====================================')
print(digits_label)
print('=====================================')
print(digits_label_names)
```

# 학습 데이터, 테스트 데이터 분리

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data, digits_label, test_size=0.2, random_state=11)
```

# 학습 데이터 세트로 학습 수행

```python
dt_clf = DecisionTreeClassifier(random_state=11)

dt_clf.fit(X_train, y_train)
```

# 테스트 데이터 세트로 예측 수행

```python
pred = dt_clf.predict(X_test)

pred
```

# 정확도 평가

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```
